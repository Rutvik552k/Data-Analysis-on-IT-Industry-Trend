```python
# 73 Coding Bootcamp;Online Courses or Certification;Other (please specify):;Colleague
def split_columns(Col_series_df):
    result_df=Col_series_df.to_frame()
    option=[]
#     idx=73 value= Coding Bootcamp;Online Courses or Certification;Other (please specify):;Colleague
    for idx,value in Col_series_df[Col_series_df.notnull()].iteritems():
#         values=["coding bootcamp",online course or certification,other specify,colllegue]
        for values in value.split(";"):
#         "coding bootcamp not present in column so if loop execute"
            if  not values in result_df.columns:
                option.append(values) #it contain the column name
                result_df[values]=False #genrate a column name and initial value is False
            result_df.at[idx,values]=True #
    return result_df[option]
```
